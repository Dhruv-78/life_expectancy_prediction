```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled= scaler.fit_transform(x)

print(x_scaled)
```

## Trying PCA and BERUTA for further Pre-processing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

X_train, X_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)

r2_scores = []

num_features = X_train.shape[1]
components_range = range(1, num_features + 1)

for n_components in components_range:
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    regressor = RandomForestRegressor(random_state=42)
    regressor.fit(X_train_pca, y_train)

    y_pred = regressor.predict(X_test_pca)

    r2 = r2_score(y_test, y_pred)
    r2_scores.append(r2)

plt.figure(figsize=(8, 6))
plt.plot(components_range, r2_scores, marker='o', linestyle='--', color='b')
plt.title("RÂ² Score vs Number of PCA Components")
plt.xlabel("Number of PCA Components")
plt.ylabel("RÂ² Score")
plt.grid(alpha=0.5)
plt.show()
```

## PCA doesn't helped much

```python
## BERUTA
from boruta import BorutaPy

rf = RandomForestRegressor(n_jobs=-1, random_state=42)

boruta = BorutaPy(estimator=rf, n_estimators='auto', random_state=42)

boruta.fit(x_scaled, y)

# Get selected features
selected_features = x.columns[boruta.support_]
print("\nSelected features by Boruta:")
print(selected_features)
```

```python
X_train = pd.DataFrame(X_train, columns=x.columns)
X_test = pd.DataFrame(X_test, columns=x.columns)

X_train_selected = X_train[['Adult_Mortality', 'Alcohol', 'Measles ', ' BMI ', 'under-five_deaths ',
       'Polio', 'Total_expenditure', ' HIV/AIDS', 'GDP', 'Population',
       ' thinness (1-19 years)', 'Income_composition_of_resources',
       'Schooling']]
X_test_selected = X_test[['Adult_Mortality', 'Alcohol', 'Measles ', ' BMI ', 'under-five_deaths ',
       'Polio', 'Total_expenditure', ' HIV/AIDS', 'GDP', 'Population',
       ' thinness (1-19 years)', 'Income_composition_of_resources',
       'Schooling']]

regressor = RandomForestRegressor(random_state=42)
regressor.fit(X_train_selected, y_train)

y_pred = regressor.predict(X_test_selected)

print(r2_score(y_test, y_pred))
```

# Model selection

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression,Ridge,PoissonRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

test_model = [LinearRegression(),Ridge(alpha=1.0),PoissonRegressor(), 
              DecisionTreeRegressor(max_depth=5),RandomForestRegressor(n_estimators=100),SVR(kernel='linear')]
model = ['Linear Regression','Ridge','Poisson Regressor','Decision tree regressor',
         'Random Forest Regressor','Support Vector Regression']
result = pd.DataFrame(columns=['Model','RMSE','r2_score'])
for i, j in enumerate(test_model):
    train_model = j.fit(X_train_selected,y_train)
    y_pred = j.predict(X_test_selected)
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    results = pd.DataFrame({'Model':[model[i]],'RMSE':[rmse],'r2_score':[r2]})
    result = pd.concat([result,results],ignore_index = True)
result
```

### Selected Models  - Random Forest , Decision tree

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

model = RandomForestRegressor(random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid_search.fit(X_train_selected, y_train)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_selected)

rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"RMSE: {rmse:.4f}")
print(f"R^2 Score: {r2:.4f}")
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

model = DecisionTreeRegressor(random_state=42)

param_grid = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5],
    'max_features': ['sqrt', 'log2', None]
}

grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error', verbose=1)
grid_search.fit(X_train_selected, y_train)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_selected)

rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"RMSE: {rmse:.4f}")
print(f"R^2 Score: {r2:.4f}")
```
